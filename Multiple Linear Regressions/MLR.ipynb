Step 4 Seperate X and Y(PROFIT)

Step 5 Build the Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,Y)
```

### Getting Coefficeint and Intercept

```python
model.intercept_
```

```python
model.coef_
```

### Profit_Pred =50122.19 + 0.8057xRND - 0.0268xADMIN + 0.0272xMKT

ADMIN and MKT do not have major impact on Profit

### Evaluate the Model

# R2 scroe

```python
model.score(X,Y)
```

Predict the result for x

```python
ypred=model.predict(X)
ypred[0:5]
```

```python
Y.head()
```

Get other Metrics

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,mean_absolute_percentage_error,r2_score
```

```python
mse=mean_squared_error(Y,ypred)
mse
```

```python
rmse=mse**(1/2)
rmse
```

```python
mae=mean_absolute_error(Y,ypred)
mae
```

```python
mape=mean_absolute_percentage_error(Y,ypred)
mape
```

with above i can say model has a good r2 score of 0.9465 >0.8 hence model is good

```python
r2=r2_score(Y,ypred)
r2
```
