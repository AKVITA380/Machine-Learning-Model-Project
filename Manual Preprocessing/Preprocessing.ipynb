fillna() - to replace missing values

```python
def replacer(df):
    cat = list(df.columns[df.dtypes=='object'])
    for i in df.columns:
        if i in cat:
            md = df[i].mode()[0]
            df[i] = df[i].fillna(md)
        else:
            mn = df[i].mean()
            df[i] = df[i].fillna(mn)
    print('Missing Values replaced in Dataframe')
```

```python

replacer(df)
```

Apply One Hot Encoding on categorical (Text) Features

```python

from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(handle_unknown='ignore',
                    sparse_output=False).set_output(transform='pandas')
X_cat_pre = ohe.fit_transform(X_cat)
X_cat_pre
```

```python

X_cat_pre.to_csv('Ohe.csv', index=False)
```

Apply Standard Scaler on Continuous (numerical) features

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().set_output(transform='pandas')
X_con_pre = scaler.fit_transform(X_con)
X_con_pre.head()
```
